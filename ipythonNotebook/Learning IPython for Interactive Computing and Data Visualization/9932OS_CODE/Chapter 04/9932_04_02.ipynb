```python
import pandas as pd
filename='data/worldcitiespop.txt'
data = pd.read_csv(filename)
```

```python
plot(data.Longitude, data.Latitude, ',')
```

```python
locations = data[['Longitude','Latitude']].as_matrix()
```

```python
def locate(position):
    d = locations - position
    distances = d[:,0] ** 2 + d[:,1] ** 2
    closest = distances.argmin()
    return data.AccentCity[closest]
print(locate(array([48.861,	 2.3358])))
```

```python
population = data['Population']
largest = population==population.max()
data.ix[largest]
```

```python
pop = array(population)
nanpop = isnan(pop)
pop[nanpop] = 0
#savez('wc/data.npz', population=pop, locations=locations)
```

```python
print(("The population count is available in {0:d} cities out of {1:d}. There are {2:.3f} billion people in "
"these cities.").format(len(pop) - nanpop.sum(), len(pop), pop[-nanpop].sum()/1e9))
```

```python
from mpl_toolkits.basemap import Basemap
m = Basemap(projection='mill',llcrnrlat=-65,urcrnrlat=85,\
            llcrnrlon=-180,urcrnrlon=180)
x0, y0 = m(-180, -65)
x1, y1 = m(180, 85)
```

```python
population = data.Population
x, y = m(locations[:,0],locations[:,1])
weights = population.copy()
weights[isnan(weights)] = 1000
h, _, _ = histogram2d(x, y, bins=(linspace(x0, x1, 500), linspace(y0, y1, 500)), weights=weights)
h[h==0] = 1
```

```python
import scipy.ndimage.filters
z = scipy.ndimage.filters.gaussian_filter(log(h.T), 1)
```

```python
figure(figsize=(10,6))
m.drawcoastlines()
m.imshow(z, origin='lower', extent=[x0,x1,y0,y1], cmap=get_cmap('Reds'))
```
